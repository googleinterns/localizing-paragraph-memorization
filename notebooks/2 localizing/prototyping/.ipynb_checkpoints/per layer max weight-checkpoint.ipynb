```python
def collect_max_gradient_weights(model, dl, c_type:str="W_in", n_batches:int=10, topK_batch_layer:int=10):
    """
    collecting the topK max gradient weights in component c_type over multiple batches
    """
    dl_l_scores, dl_l_idcs = [],[]
    for batch_i in tqdm.tqdm(range(n_batches)):
        c_toks_NI, k_toks_NI = next(dl)
        fwd_cache, bwd_cache = gradient.run_fwd_bwd(model, c_toks_NI, k_toks_NI)
        c_vals, c_names = localizing.collect_c_type(model=model, cache=None, c_type=c_type)
        l_scores, l_idcs = torch.topk(c_vals.view(c_vals.shape[0], -1), topK_batch_layer, largest=True, dim=-1)
        dl_l_scores.append(l_scores)
        dl_l_idcs.append(l_idcs)
    return dl_l_scores, dl_l_idcs


def extract_max_weights_per_layer(scores_bl:list, idcs_bl:list, min_count:int= 1, sort_by:str="score"):
    """
    extract the top weights of a component per layer and return their index, mean score and count
    """
    ## reshaping to remove batch dimension_____________________________
    scores_l = torch.stack(scores_bl).transpose(0,1)
    n_batches = scores_l.shape[1]
    scores_l = scores_l.reshape(scores_l.shape[0], -1)
    idcs_l = torch.stack(idcs_bl).transpose(0,1)
    idcs_l = idcs_l.reshape(idcs_l.shape[0], -1)

    ## gathering_______________________________________________________
    max_ids_lk, max_counts_lk, max_scores_lk = [], [], []
    for l, (scores, idcs) in enumerate(tqdm.tqdm(zip(scores_l, idcs_l))):

        weight_ids, weight_idcs, weight_counts = torch.unique(idcs, return_inverse=True, return_counts=True, dim=-1)
        weight_scores = torch.bincount(weight_idcs, weights=scores) / torch.bincount(weight_idcs) ##mean score
        weight_ids, weight_counts, weight_scores = weight_ids[weight_counts >= min_count], weight_counts[weight_counts >= min_count], weight_scores[weight_counts >= min_count]

        if sort_by=="score": ## sort by score
            sort_idcs = torch.argsort(weight_scores, dim=-1, descending=True) 
        elif sort_by=="count": ## sort by count
            sort_idcs = torch.argsort(weight_counts, dim=-1, descending=True) 
            
        weight_ids, weight_counts, weight_scores =  weight_ids[sort_idcs], weight_counts[sort_idcs], weight_scores[sort_idcs]
        max_ids_lk.append(weight_ids)
        max_counts_lk.append(weight_counts)
        max_scores_lk.append(weight_scores)
    return max_ids_lk, max_counts_lk, max_scores_lk
```

```python
scores_bl, idcs_bl = collect_max_gradient_weights(model, dl, c_type="W_in", n_batches=3, topK_batch_layer=10)
max_ids_lk, max_counts_lk, max_scores_lk = extract_max_weights_per_layer(scores_bl, idcs_bl, min_count=1, sort_by="score")
```

```python
c_type = "W_in"
localizing.store_data((max_ids_lk, max_counts_lk, max_scores_lk), f"results/{model_type}/{c_type}.pt")
```

```python
def plot_max_layer_weights(max_ids_lk:list, max_counts_lk:list, max_scores_lk:list, n_batches:int=None, c_type:str="W_in"):
    
    ## plotting__________________________________
    n_layers = len(max_ids_lk)  
    if n_batches is None:
        n_batches = torch.max(max_counts_lk[0]).item()
    
    fig, axs = plt.subplots(n_layers, 1, figsize=(9, 11), gridspec_kw={'hspace': 0.1, 'wspace': 0.1}, constrained_layout=True)
    fontsize=10

    for l, (ids_k, counts_k, scores_k) in enumerate(zip(max_ids_lk, max_counts_lk, max_scores_lk)):

        ids_k = ids_k.detach().tolist()
        x = np.arange(len(ids_k))
        counts_k = counts_k.detach().tolist()
        scores_k = scores_k.detach().tolist()
        sns.barplot(x=x, y=counts_k, ax=axs[l], color='b')
        axs[l].hlines(n_batches, xmin=-0.5, xmax=len(ids_k), color='b', linestyles=":")
        axs[l].set_ylabel('count', color='b', fontsize=fontsize)

        ax2 = axs[l].twinx()
        ax2.scatter(x, scores_k, color='r', marker='o')
        ax2.set_ylabel('value', color='r', fontsize=fontsize)
        ax2.set_ylim(0, None)

        corr = scipy.stats.kendalltau(scores_k, counts_k).statistic
        axs[l].set_title(f"{c_type}, layer {l}, corr {round(corr,2)}",loc='left',fontsize=fontsize)  
        axs[l].set_xticks(x, labels=ids_k, fontsize=fontsize-5, rotation=20)
    #fig.savefig("W_in max neurons.png", dpi=200, bbox_inches='tight')
    
plot_max_layer_weights(max_ids_lk, max_counts_lk, max_scores_lk)
```
